```python
import matplotlib.pyplot as plt

# Select the GPP and NDVI values at the specified location (latitude: 43.875, longitude: -93.875)
gpp_time_series = ds_filtered['gross_primary_productivity'].sel(
    lat=43.875, 
    lon=-93.875, 
    method="nearest"  # Ensure selection even if exact lat/lon aren't present
)

# Select NDVI (assuming it's in the dataset with the variable name 'ndvi')
ndvi_time_series = ds_filtered['ndvi'].sel(
    lat=43.875, 
    lon=-93.875, 
    method="nearest"  # Ensure selection even if exact lat/lon aren't present
)

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot GPP on the left axis
ax1.plot(gpp_time_series.time, gpp_time_series, label="GPP", color='green')
ax1.set_xlabel("Time")
ax1.set_ylabel("GPP (g C m$^{-2}$ d$^{-1}$)", color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Create a second y-axis for NDVI
ax2 = ax1.twinx()
ax2.plot(ndvi_time_series.time, ndvi_time_series, label="NDVI", color='blue')
ax2.set_ylabel("NDVI", color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title and grid
plt.title("Time Series of GPP and NDVI at Latitude 43.875, Longitude -93.875 (2001-2020)")
ax1.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np

# Make sure the time series data is aligned by time (both GPP and NDVI have the same time axis)
gpp_values = gpp_time_series.values
ndvi_values = ndvi_time_series.values

# Calculate the Pearson correlation coefficient
correlation = np.corrcoef(gpp_values, ndvi_values)[0, 1]

# Print the correlation value
print(f"Pearson's correlation coefficient between GPP and NDVI: {correlation}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Assuming gpp_time_series and ndvi_time_series are already defined

# 1. Calculate the 5th and 95th percentiles for GPP and NDVI
gpp_5th_percentile = gpp_time_series.quantile(0.05)
gpp_95th_percentile = gpp_time_series.quantile(0.95)

ndvi_5th_percentile = ndvi_time_series.quantile(0.05)
ndvi_95th_percentile = ndvi_time_series.quantile(0.95)

print(f"GPP 5th percentile: {gpp_5th_percentile}, GPP 95th percentile: {gpp_95th_percentile}")
print(f"NDVI 5th percentile: {ndvi_5th_percentile}, NDVI 95th percentile: {ndvi_95th_percentile}")

# 2. Normalize GPP and NDVI using the percentiles
gpp_normalized = (gpp_time_series - gpp_5th_percentile) / (gpp_95th_percentile - gpp_5th_percentile)
ndvi_normalized = (ndvi_time_series - ndvi_5th_percentile) / (ndvi_95th_percentile - ndvi_5th_percentile)

# Clip values to the range [0, 1]
gpp_normalized = np.clip(gpp_normalized, 0, 1)
ndvi_normalized = np.clip(ndvi_normalized, 0, 1)

# 3. Compute the EQI as the average of normalized GPP and NDVI
eqi = (gpp_normalized + ndvi_normalized) / 2

# 4. Plot the EQI time series for the location (43.875, -93.875)
plt.figure(figsize=(10, 6))
plt.plot(gpp_time_series.time, eqi, label="EQI", color="green")
plt.title("Time Series of EQI at Latitude 43.875, Longitude -93.875 (2001-2020)")
plt.xlabel("Time")
plt.ylabel("EQI")
plt.grid(True)
plt.legend()
plt.show()

# 5. Classify EQI into different quality levels
eqi_classification = np.where(eqi < 0.3, "Low", np.where(eqi < 0.7, "Medium", "High"))

# Add classification labels to the plot
plt.figure(figsize=(10, 6))
plt.plot(gpp_time_series.time, eqi, label="EQI", color="green")
plt.scatter(gpp_time_series.time, eqi, c=eqi_classification, cmap="viridis", label="Quality Level")
plt.title("Time Series of EQI with Classification at Latitude 43.875, Longitude -93.875")
plt.xlabel("Time")
plt.ylabel("EQI")
plt.grid(True)
plt.legend()
plt.show()

# 6. Calculate the trend of EQI using a simple linear regression
slope, intercept, r_value, p_value, std_err = linregress(np.arange(len(eqi)), eqi)

# Plot the EQI trend line
plt.figure(figsize=(10, 6))
plt.plot(gpp_time_series.time, eqi, label="EQI", color="green")
plt.plot(gpp_time_series.time, slope * np.arange(len(eqi)) + intercept, label="Trend Line", color="red", linestyle="--")
plt.title("Time Series of EQI with Trend Line at Latitude 43.875, Longitude -93.875")
plt.xlabel("Time")
plt.ylabel("EQI")
plt.grid(True)
plt.legend()
plt.show()

print(f"EQI Trend slope: {slope}")

# 7. Create a spatial map of EQI for the first time step (or any other specific time step)
eqi_first_time_step = eqi[0]  # Assuming eqi is a 1D array (time, lat, lon)

# Assuming ds_filtered contains the spatial data
plt.figure(figsize=(12, 8))
plt.pcolormesh(ds_filtered['lon'], ds_filtered['lat'], eqi_first_time_step, cmap="viridis")
plt.colorbar(label="EQI")
plt.title("EQI Spatial Map at Time Step 1")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
